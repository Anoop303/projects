```python
import numpy as np
```

```python
altitude= np.random.uniform(0,10000,1000)
velocity =np.random.uniform(0,400,1000)
acceleration =np.random.uniform(-20,20,1000)
```

```python
critical_altitude=altitude[altitude>5000]
critical_velocity=velocity[velocity>300]
critical_acceleration=acceleration[acceleration<-9.8]
```

```python
critical_values=np.concatenate((critical_altitude,critical_velocity,critical_acceleration))
```
